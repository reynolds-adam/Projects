Building a machine learning model can be a complicated process. Many new data scientists like myself have spent countless hours wondering how to get started. You may be familiar with several machine learning algorithms, but do not know how to apply them in practice.

Fortunately, we can break the process down into steps which make it easier to digest. 

1. Look at the big picture
2. Get the data
3. Explore and visualize the data
4. Prepare the data for maching learning algorithms
5. Select a model and train it
6. Fine tune your model
7. Present your solution
8. Launch, monitor and maintain your system

If you have spent much time doing data ananlysis, you are probably familiar with at least a few of these steps. The goal of this article is to give a basic overview of the steps involved in building machine learning models.

I will be using the Mall Customer Segmentation Dataset from Kaggle.

# Get the Data

```python
# Import needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the data
from sklearn.datasets import load_diabetes
df = pd.read_csv('Mall_Customers.csv')
# df = pd.DataFrame(df.data, columns=df.feature_names)
```

- Familiarize yourself with the schema of the data

# Explore and visualize the data
- Look at the data structure

```python
# look at the first 5 rows
df.head(5)
```

```python
# gives a desctiption of the data and helps identify any null values
df.info()
```

```python
# Explore Categorical Fields
df['Gender'].value_counts()
```

```python
# Explore Numerical Fields
df.describe()
```

```python
# Plot histograms of all numerical fields
df.hist(bins = 20, figsize=(20,15))
plt.show()
```

After plotting histograms, look for distributions that are tail-heavy. You may need to transform such distributions in order have more of a Gaussian distribution.

```python
# Separate data into Train and Test Sets
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=.2, random_state=42)
```

Now we will explore the train set. Do not look at the test set until your model is complete.

```python
# Look for any correlations in the data
corr_matrix=train_set.corr()
# Look at correlation of all fields to a particular field
corr_matrix['Age'].sort_values(ascending=False)
```

```python
# Plot a scatter matrix to see correlations
attributes = ['Age','Annual Income (k$)','Spending Score (1-100)']
sns.set()
pd.plotting.scatter_matrix(df[attributes],figsize=(12,8))
plt.show()
```
